```python
import requests
import pandas as pd

api_key = '18d193d2c6afaa46ff71f30e9480554d'

def get_walk_score(lat, lon, address):
    try:
        url = f"https://api.walkscore.com/score?format=json&address={address}&lat={lat}&lon={lon}&wsapikey={api_key}"
        response = requests.get(url)
        if response.status_code == 200:
            return response.json().get('walkscore')
        else:
            # Log error or take appropriate action
            return None
    except Exception as e:
        # Log exception details or take appropriate action
        return None

# Update the DataFrame with Walk Scores
for index, row in zillow_df.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']) and pd.notna(row['Address']):
        walk_score = get_walk_score(row['Latitude'], row['Longitude'], row['Address'])
        if walk_score is not None:
            zillow_df.at[index, 'Walk Score'] = walk_score

# Check the updated DataFrame
zillow_df.head()
```
